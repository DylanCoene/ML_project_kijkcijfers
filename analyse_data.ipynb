```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Importeren van de dataset

```python
kijkcijfers = pd.read_csv('./data/tv_kijkcijfers_start.csv')
kijkcijfers.sample(5)
```

# Korte data exploration + notities

```python
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

# Histogrammen genereren van dataset
kijkcijfers.hist(bins=50, figsize=(12,8))
plt.show()
```

```python
kijkcijfers['log_viewers'] = kijkcijfers['viewers'].apply(lambda x: 0 if x == 0 else np.log(x))
kijkcijfers['sqrt_viewers'] = kijkcijfers['viewers'].apply(lambda x: np.sqrt(x))
```

```python
kijkcijfers.hist(bins=50, figsize=(12,8))
plt.show()
```
